```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head() 
```

```python
df.describe() 
```

```python
sns.distplot(df['price'],bins=10) 
```

```python
sns.distplot(df['price'],kde=False,bins=20)      
```

```python
sns.jointplot(x='balcony',y='price',data=df,kind='scatter',xlim=(0,3),ylim=(0,3000))
```

```python
sns.pairplot(df) 
```

```python
sns.kdeplot(df['price']) 
sns.rugplot(df['price'])  
```

```python
sns.pairplot(df,hue='bath',palette='coolwarm') 
```

```python
sns.jointplot(x='balcony',y='price',data=df,kind='kde')   
```

```python
sns.barplot(x='balcony',y='price',data=df,estimator=np.std)  
```

```python
sns.countplot(x='bath',data=df)    
```

```python
sns.boxplot(x='area_type',y='price',data=df,palette='rainbow') 
figsize = (100,200)  
```

```python
sns.boxplot(data=df,palette='rainbow',orient='h')  
```

```python
sns.boxplot(x='balcony',y='price',hue='area_type',palette='coolwarm',data=df) 
```

```python
sns.violinplot(x='balcony',y='price',data=df,hue='area_type',palette='Set1') 
```

```python
sns.stripplot(x='balcony',y='price',data=df,jitter=True,hue='area_type')     
```

```python
sns.factorplot(x='balcony',y='price',data=df,hue='area_type',kind='bar') 
 
```
